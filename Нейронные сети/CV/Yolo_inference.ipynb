```python
from ultralytics import YOLO
import matplotlib.pyplot as plt
import cv2
```

```python
# Загрузка обученной модели
model = YOLO("yolov8m.pt")

# Использование на изображении
results = model('/content/bangkok.jpg')

# Вывод результатов распознавания
for result in results:
    boxes = result.boxes
    result.show()  # отображение обнаруженных объектов
    result.save(filename="result.jpg")  # сохранение
```

```python
# Отображение изображения с ограничивающими рамками
image = cv2.imread("/content/result.jpg")
plt.figure(figsize=(20, 15))
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```
